# Exploración y Limpieza de Datos - Airbnb

En este notebook realizaremos la carga de los datos crudos, la limpieza de valores nulos y el análisis inicial del mercado.

```python
import pandas as pd
import sys
import os

# Añadir el directorio src al path para importar nuestros módulos
sys.path.append(os.path.abspath('../src'))
from cleaning import clean_airbnb_data
```

### 1. Carga de Datos

```python
# Cargamos el archivo de la ciudad actual
df_raw = pd.read_csv('../data/raw/listings.csv')
print(f"Dimensiones originales: {df_raw.shape}")
df_raw.head()
```

### 2. Limpieza de Datos

Aplicamos el pipeline de limpieza definido en `src/cleaning.py` que:
- Elimina duplicados.
- Limpia la columna de precios y elimina filas con precios nulos.
- Corrige métricas de reseñas.
- Normaliza textos y elimina columnas 100% vacías.

```python
df_clean = clean_airbnb_data(df_raw)
print(f"Dimensiones finales: {df_clean.shape}")
```

### 3. Verificación de Nulos

```python
print("Valores nulos por columna:")
print(df_clean.isnull().sum())

print(f"\nPrecio Promedio Final: ${df_clean['price'].mean():.2f}")
```

### 4. Exportar Datos

Guardamos el dataset limpio para realizar el análisis profundo (EDA) en el siguiente notebook.

```python
# Crear carpeta processed si no existe
os.makedirs('../data/processed', exist_ok=True)

# Guardar
df_clean.to_csv('../data/processed/listings_clean.csv', index=False)
print("✅ Datos guardados en data/processed/listings_clean.csv")
```
